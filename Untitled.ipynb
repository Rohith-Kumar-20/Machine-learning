```python
import string
string.punctuation
```

```python
def remove_punctuation(text):
    punctuationfree = "".join(i for i in text if i not in string.punctuation)
    return punctuationfree
```

```python
stopwords= nltk.corpus.stopwords.words('english')
stopwords
```

```python
def remove_stopwords(text):
    output = " ".join(word for word in text.split() if word.lower() not in stopwords )
    return output
```

```python
df1['posts']=df1['posts'].apply(lambda x:remove_stopwords(x))
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizer(text):
    lemm_text = " ".join([wordnet_lemmatizer.lemmatize(word) for word in text.split()])
    return lemm_text
```

```python
df1['posts'] = df1['posts'].apply(lambda x:lemmatizer(x))
```

```python
df1['posts']
```

```python
def remove_digits(text):
    clean_text= re.sub('[^a-z]+',' ',text)
    return clean_text
```

```python
df1['posts']=df1['posts'].apply(lambda x:remove_digits(x))
```
